```python
import torch
import numpy as np

seq_len = 100
weights = 5

# Generating the tensor
times = torch.tensor(
    np.transpose(np.tile(range(seq_len), (weights, 1))),
    device='cpu'  # You might specify a different device here
)

print(times)
```

```python
cutoff_times = torch.ones(weights) * -1
computed = [seq_len, seq_len-1, -1, 0, 1]
cutoff_times = torch.tensor(computed)

cutoff_times
```

```python
diff = times - cutoff_times
diff
```

```python
stuff = (diff) / 0.1
stuff
```

```python
weight_vector = torch.round(torch.sigmoid(stuff), decimals=2)#.reshape(1, seq_len, weights)
weight_vector
```

```python
import matplotlib.pyplot as plt

for i in range(weight_vector.size(1)):
    plt.plot(weight_vector[:, i], marker="o")
    plt.title("Cutoff value = " + str(cutoff_times[i].item()))
    plt.ylim(-0.1, 1.1)
    plt.yticks([0,0.5,1])
    plt.grid()
    plt.show()
```
